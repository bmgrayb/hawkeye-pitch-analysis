```python
import os
from os import listdir
from os.path import isfile, join
import pandas as pd
pd.options.mode.chained_assignment = None
from pitch_path.processing.data_processing import PitcherDataProcessor
import logging
import sys
logger = logging.getLogger()
logging.basicConfig(stream=sys.stdout, level=logging.INFO)

```

## Save off processed data

```python
data = "../data"
inputs = f"{data}/input"
pitcher_files = [join(inputs, f) for f in listdir(inputs) if isfile(join(inputs, f))]
```

```python

for f in pitcher_files:
    PitcherDataProcessor(file_name=f).save_pitcher_df("../data/output")
```

## Save off features

```python
data = "../data"
inputs_processed = f"{data}/output/processed"
pitcher_files_processed = [join(inputs_processed, f) for f in listdir(inputs_processed) if isfile(join(inputs_processed, f))]
```

```python
for f in pitcher_files_processed:
    PitcherDataProcessor(file_name=f, is_processed_file=True).save_pitcher_features_df("../data/output")
```

## Training Data Set

```python
data = "../data"
features = f"{data}/output/pitcher_features"
pitcher_features = [join(features, f) for f in listdir(features) if isfile(join(features, f))]
output_dir = f"{data}/output/training/"
output_file_path = f"{output_dir}/training_data"
```

```python
all_features_list = []
for f in pitcher_features:
    all_features_list.append(pd.read_feather(f))
```

```python
all_features = pd.concat(all_features_list)
```

```python
all_features.head()
```

```python
if not os.path.exists(output_dir):
            os.mkdir(output_dir)
all_features.to_feather(path=output_file_path)
```
